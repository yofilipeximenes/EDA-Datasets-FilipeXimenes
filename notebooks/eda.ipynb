# 2. Exploring the dataset


En esta fase, realizamos una auditoría de calidad de los datos para identificar tres problemas críticos antes de cualquier análisis profundo:

1.  **Valores Nulos (Missing Values):** Cuantificar el porcentaje de datos faltantes por columna para decidir si imputar (rellenar) o eliminar registros.
2.  **Duplicados:** Detectar redundancias que puedan sesgar los conteos de frecuencia.
3.  **Cardinalidad:** Entender cuántos valores únicos existen en variables clave (ej. cuántos compositores distintos hay).

```python
# Función auxiliar para auditoría de nulos
def auditoria_nulos(df, nombre_df):
    nulos = df.isnull().sum()
    total = len(df)
    porcentaje = (nulos / total) * 100
    
    # Creamos un DataFrame resumen
    reporte = pd.DataFrame({
        'Valores Nulos': nulos,
        'Porcentaje (%)': porcentaje,
        'Tipo de Dato': df.dtypes
    })
    
    # Filtramos solo las columnas que tienen nulos y ordenamos
    reporte = reporte[reporte['Valores Nulos'] > 0].sort_values(by='Porcentaje (%)', ascending=False)
    
    print(f"\n--- REPORTE DE NULOS: {nombre_df} ---")
    if reporte.empty:
        print("¡Integridad perfecta! No se detectaron valores nulos.")
    else:
        display(reporte)
        
# Ejecutamos la auditoría en ambas tablas
auditoria_nulos(df_main, "DF_MAIN (Conciertos)")
auditoria_nulos(df_soloists, "DF_SOLOISTS (Solistas)")
```

#### Verificación de Claves para Fusión (Join Keys)

Para relacionar los solistas con las obras y fechas, es necesario identificar la clave común (Foreign Key). Se inspeccionan las columnas de identificadores para asegurar que existen en ambas tablas.

```python
# Buscamos columnas comunes que sirvan de ID
interseccion_columnas = set(df_main.columns).intersection(set(df_soloists.columns))

print(f"Columnas comunes entre ambos datasets: {interseccion_columnas}")

# Inspección rápida de los IDs para ver si tienen el mismo formato
# (Ajusta 'id' o 'programID' según lo que salga en la intersección)
columnas_id = [col for col in interseccion_columnas if 'id' in col.lower() or 'program' in col.lower()]

if columnas_id:
    col_clave = columnas_id[0] # Tomamos la primera coincidencia como candidata
    print(f"\nPosible clave de unión detectada: '{col_clave}'")
    print(f"Ejemplo IDs en Main: {df_main[col_clave].head(3).tolist()}")
    print(f"Ejemplo IDs en Solistas: {df_soloists[col_clave].head(3).tolist()}")
else:
    print("\n⚠️ ALERTA: No se detectaron columnas obvias de ID en común por nombre.")
```

# 3.3 Limpieza y Normalización

En base a la auditoría de la sección anterior, procedemos a ejecutar las acciones correctivas necesarias para garantizar la calidad del análisis.

### 3.3.1 Tratamiento de Duplicados
Se ha detectado una redundancia crítica en el conjunto de datos principal (`df_main`) con un **61.78%** de filas duplicadas. Esto sugiere una desnormalización excesiva o errores en la captura de datos.
* **Decisión:** Eliminar los duplicados exactos para evitar sesgar los conteos estadísticos (ej. contar el mismo concierto tres veces).

### 3.3.2 Homogeneización de Claves (Join Key)
Para permitir la integración de datos, se ha identificado `programID` como la clave común. Sin embargo, existe una discrepancia de tipos:
* `df_main`: Formato numérico flotante (`float`).
* `df_soloists`: Formato entero (`int`).
* **Acción:** Convertir la clave en `df_main` a entero para permitir cruces exactos.

### 3.3.3 Normalización de Texto
Se estandarizarán las cadenas de texto (instrumentos, compositores) a minúsculas y sin espacios superfluos para evitar categorías fantasmas (ej. "Piano" vs "piano ").

```python
# --- 1. ELIMINACIÓN DE DUPLICADOS ---
print(f"Filas en df_main antes de limpieza: {len(df_main)}")
df_main = df_main.drop_duplicates()
print(f"Filas en df_main tras eliminar duplicados: {len(df_main)}")

# Hacemos lo mismo con solistas por higiene
df_soloists = df_soloists.drop_duplicates()

# --- 2. CORRECCIÓN DE TIPOS EN CLAVE (programID) ---
# Primero rellenamos nulos en el ID con 0 (si los hubiera) para poder convertir a int
df_main['programID'] = df_main['programID'].fillna(0).astype(int)

# Verificamos que ahora ambos son compatibles
print("\n--- Verificación de Claves ---")
print(f"Tipo de programID en Main: {df_main['programID'].dtype}")
print(f"Tipo de programID en Solistas: {df_soloists['programID'].dtype}")

# --- 3. MERGE (FUSIÓN DE TABLAS) ---
# Dado que df_main ya parece tener columnas de solistas (vimos 'soloistName' en intersección),
# vamos a comprobar si df_main ya tiene la info completa o si necesitamos el merge.
# Si df_main ya tiene 'soloistInstrument' lleno, el merge es redundante.

nulos_solista_main = df_main['soloistName'].isnull().sum()
print(f"\nNulos en 'soloistName' dentro de df_main: {nulos_solista_main}")

# ESTRATEGIA DE FUSIÓN:
# Si el dataset principal ya tiene la info, trabajamos con él. 
# Si no, haríamos: df_completo = pd.merge(df_main, df_soloists, on='programID', how='left')
# Para este ejercicio, asumiremos que df_main limpio es nuestra base principal.
```

```python
# --- 4. NORMALIZACIÓN DE CATEGORÍAS ---

# Lista de columnas de texto a limpiar en df_main
cols_texto = ['composerName', 'workTitle', 'conductorName', 'orchestra']

# Aplicamos limpieza solo a las columnas que existen
for col in cols_texto:
    if col in df_main.columns:
        # Convertir a string, pasar a minúsculas y quitar espacios
        df_main[col] = df_main[col].astype(str).str.lower().str.strip()

# Limpieza específica de fechas (Crear columnas derivadas para análisis temporal)
# Esto es esencial para el gráfico de "Evolución Temporal"
df_main['Year'] = df_main['Date'].dt.year
df_main['Month'] = df_main['Date'].dt.month

print("--- Muestra de datos normalizados ---")
display(df_main[['composerName', 'workTitle', 'Year']].head())
```

# 4. Visualization

## 1. Visualizaciones Básicas

La visualización es el paso final del EDA. Permite confirmar los hallazgos de limpieza y extraer los patrones iniciales del dataset. Se utilizará Matplotlib/Seaborn para generar las tres visualizaciones obligatorias.

Se analizará:
1.  **Distribución Temporal:** Tendencia de conciertos por año (Histograma).
2.  **Frecuencia Categórica:** Los compositores más interpretados (Gráfico de barras).
3.  **Patrón Relevante:** Distribución de conciertos a lo largo del tiempo (Línea o Puntos).

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

# Histograma de la variable 'Year' (creada en la limpieza)
# Esto muestra la distribución de la frecuencia de conciertos a lo largo del tiempo.
sns.histplot(df_main['Year'], bins=50, kde=False)

plt.title('Distribución de Conciertos de la Filarmónica por Año (1842 - Actualidad)')
plt.xlabel('Año del Concierto')
plt.ylabel('Conteo de Eventos')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Contamos los 10 compositores más frecuentes tras la normalización
top_compositores = df_main['composerName'].value_counts().head(10).sort_values(ascending=False)

plt.figure(figsize=(14, 7))

# Gráfico de barras
sns.barplot(x=top_compositores.index, y=top_compositores.values, palette="viridis")

plt.title('Top 10 Compositores Más Interpretados (Recuento de Obras)')
plt.xlabel('Compositor (Normalizado)')
plt.ylabel('Frecuencia Absoluta')
plt.xticks(rotation=45, ha='right') # Rotamos etiquetas para legibilidad
plt.tight_layout()
plt.show()
```

```python
# Agrupamos por año para ver la tendencia (Esto es más informativo que el simple histograma)
conciertos_por_año = df_main.groupby('Year').size()

plt.figure(figsize=(15, 6))

# Gráfico de línea para mostrar la evolución temporal
conciertos_por_año.plot(kind='line', marker='o', markersize=3, color='darkred')

plt.title('Evolución Histórica del Número de Eventos Anuales')
plt.xlabel('Año')
plt.ylabel('Número de Eventos (Obras)')
plt.grid(axis='both', linestyle='--', alpha=0.6)
plt.ylim(bottom=0)
plt.show()
```
